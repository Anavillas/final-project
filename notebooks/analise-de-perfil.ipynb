#### Verificando se precisa de Limpeza em cada tabela

Tabela de Clientes

```python
df_clientes.head(100)
```

Limpeza nos prefixos dos nomes, e retirar espaços em branco ao final e começo.

```python
df_clientes['nome'] = df_clientes['nome'].str.removeprefix('Dr.')
df_clientes['nome'] = df_clientes['nome'].str.removeprefix('Sra.')
df_clientes['nome'] = df_clientes['nome'].str.removeprefix('Sr.')
df_clientes['nome'] = df_clientes['nome'].str.removeprefix('Dra.')

df_clientes['nome'] = df_clientes['nome'].str.strip()
df_clientes.head(100)
```

Convertendo em número a coluna de data
O máximo de dias de contrato, são 730 dias, 2 anos.

```python
df_contratos['data_inicio'] = pd.to_datetime(df_contratos['data_inicio'])
df_contratos['data_fim'] = pd.to_datetime(df_contratos['data_fim'])
df_contratos['periodo contrato'] = (df_contratos['data_fim'] - df_contratos['data_inicio']).dt.days

df_contratos['periodo contrato'].max()
```
